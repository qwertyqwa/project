```python
import cianparser
import time
import pandas as pd

def save_data_to_csv(data, filename="cian_data.csv", mode="a"):
    """Функция записи данных в CSV-файл.

    Args:
        data (list): Список словарей с данными.
        filename (str, optional): Имя файла. Defaults to "cian_data.csv".
        mode (str, optional): Режим записи. Defaults to "a" (append).
    """
    df = pd.DataFrame(data, index=range(len(data))) # Добавили индекс
    df.to_csv(filename, index=False, header=False, mode=mode)

# Создаем парсер для обоих регионов
parser = cianparser.CianParser(location="Моска и Московская область") 

# Записываем заголовки в файл
save_data_to_csv(parser.get_flats(deal_type="sale", rooms=(1, 2), with_saving_csv=False, additional_settings={"start_page":1, "end_page":1})[0], filename="cian_data.csv", mode="w")

data = []

# Парсим данные по 345 страницам для обоих регионов
for page in range(2, 400):
    time.sleep(40)
    page_data = parser.get_flats(deal_type="sale", rooms=(1, 2), with_saving_csv=False, additional_settings={"start_page":page, "end_page":page})
    data.extend(page_data)
    save_data_to_csv(page_data, filename="cian_data.csv")

print("Данные успешно сохранены в файл cian_data.csv")
```
